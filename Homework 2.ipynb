## Question 2.5

```python
df2KO = pd.read_csv("2.5KO.csv")
df2PEP = pd.read_csv("2.5PEP.csv")
df2KO.head()
```

```python
df2PEP.head()
```

```python
df2a = pd.concat([df2KO["Open"], df2PEP["Open"]], axis=1, join='inner')
df2a.corr()
KO_PEP_corr = 0.893211
print(f"The correlation coefficient between the stock prices of these two corporations is {KO_PEP_corr}")
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df2KO["Open"], df2PEP["Open"])
plt.xlabel("KO stock price")
plt.ylabel("PEP stock price")
plt.title("KO-PEP stock price comparison")
plt.savefig("figure/2.5 KO-PEP.png")
```

```python
KO_values = np.array([np.min(df2KO["Open"]), np.max(df2KO["Open"])])
PEP_values = np.array([np.min(df2PEP["Open"]), np.max(df2PEP["Open"])])
KO_values_standardized = (KO_values - df2KO["Open"].mean()) / df2KO["Open"].std(ddof=0)

PEP_values_predicted = KO_values_standardized * KO_PEP_corr * df2PEP["Open"].std(ddof=0) + df2PEP["Open"].mean()

plt.figure(figsize=(8,8))
plt.scatter(df2KO["Open"], df2PEP["Open"])
plt.xlabel("KO stock price")
plt.ylabel("PEP stock price")
plt.title("KO-PEP stock price comparison")
plt.savefig("figure/2.5 KO-PEP.png")
plt.plot(KO_values, PEP_values_predicted, "r", linewidth=5)
plt.savefig("figure/2.5 ko pep corr.png")
```
