```python
# Library import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
df = pd.read_csv("wine.data", dtype=np.float32)
df.columns = ["Region", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", \
"Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", \
"Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"]
df
```

```python
region_one_flav = df.loc[df["Region"] == 1, ["Flavanoids"]].to_records()
N1 = len(region_one_flav)
r1f_mean = np.mean([s[1] for s in region_one_flav])
r1f_stdunbised = np.sqrt(1/(N1-1) * np.sum([(s[1] - r1f_mean) ** 2 for s in region_one_flav]))
r1f_stderr = r1f_stdunbised / np.sqrt(N1)
print("The 99% confidence interval of the mean value of the flavanoids variable for wine from region 1 is:")
print(f"{r1f_mean - 2.55 * r1f_stderr} to {r1f_mean + 2.55 * r1f_stderr}")
```

```python
region_three_flav = df.loc[df["Region"] == 3, ["Flavanoids"]].to_records()
N3 = len(region_three_flav)
r3f_mean = np.mean([s[1] for s in region_three_flav])
r3f_stdunbised = np.sqrt(1/(N3-1) * np.sum([(s[1] - r3f_mean) ** 2 for s in region_three_flav]))
r3f_stderr = r3f_stdunbised / np.sqrt(N3)
print("The 99% confidence interval of the mean value of the flavanoids variable for wine from region 1 is:")
print(f"{r3f_mean - 2.55 * r3f_stderr} to {r3f_mean + 2.55 * r3f_stderr}")
```
