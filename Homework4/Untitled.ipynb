```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from jupyterthemes import jtplot
jtplot.style(theme="gruvboxd")
```

```python
results = np.zeros((10, 2))
numTickets = 12
numSeats = 10

for i, p in enumerate(np.linspace(0.1, 1.0, num=10)):
    arrivals = np.random.random((numTickets, 100000)) < p
    numArrivals = arrivals.sum(axis=0)
    women = (np.random.random((numTickets, 100000)) >= 0.5) * arrivals
    num = women.sum(axis=0)
    results[i] = [p, numArrivals[num >= 2].mean()]
results
```

```python
plt.plot(results[:,0], results[:,1], 'bo-')
plt.xlabel('Probability of arrival p')
plt.ylabel('Expected value')
plt.title('Expected number of ticket holders that show up given that there are at least two women')
```
