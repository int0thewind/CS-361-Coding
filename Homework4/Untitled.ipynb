```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from jupyterthemes import jtplot
jtplot.style(theme="gruvboxl")
```

```python
results = np.zeros((10, 2))
tickets = 10
totalTrials = 100000
for i, p in enumerate(np.linspace(0.1, 1.0, num=10)):
    passengersOnPlane = np.random.random((tickets, totalTrials)) < p
    numberOfArrivals = passengersOnPlane.sum(axis=0)
    womenInPassengers = np.random.random((tickets, totalTrials)) >= 0.5
    womenAndOnPlane = passengersOnPlane * womenInPassengers
    numberOfWomen = womenAndOnPlane.sum(axis=0)
    womenCondition = numberOfWomen >= 2
    results[i] = [p, numberOfArrivals[womenCondition].mean()]
```

```python
plt.figure(figsize=(7, 5))
plt.plot(results[:,0], results[:,1], 'bo-')
plt.xlabel('Probability of arrival p')
plt.ylabel('Expected value')
plt.title('Expected number of ticket holders that show up given that there are at least two women')
```
